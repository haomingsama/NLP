```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, brier_score_loss, fbeta_score
from sklearn.metrics import auc, roc_curve, roc_auc_score, precision_recall_curve
```

```python
data = pd.read_csv('spambase.data').as_matrix()
train_split = int(data.shape[0]*0.2)
print('有%s行数据，%s个特征'%(data.shape[0],data.shape[1]))
print('标注为spam 的数据有 %s%%   \n标注为ham 的数据有 %s%% '%(round(100*np.sum(data[:,-1])/4600,2), round(100*(1-np.sum(data[:,-1])/4600),2)))
print('按照经验法，训练集要有 %s 个数据，'%(train_split))
```

```python
# 想要每次运行程序都将不同的数据分进training set 和test set 所以这里要打乱数据
np.random.shuffle(data)
```

```python
# 前48行都是有关于word 的 term frequency 数据
x = data[:,:57]
# 最后一个column是label 数据
y = data[:,-1]

#分割训练集和测试集
x_train = x[:-train_split,]
y_train = y[:-train_split,]
x_test = x[-train_split:,]
y_test = y[-train_split:,]
```

### MultinomialNB

```python

model = MultinomialNB()
model.fit(x_train,y_train)
print('Classfication rate for Naive Bayes %s'%(model.score(x_test,y_test)) )
model_report('MultinomialNB',model)
```

### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
model.fit(x_train,y_train)
model_report('AdaBoost',model)
print('Classfication rate for AdaBoost %s'%(model.score(x_test,y_test)) )
```
