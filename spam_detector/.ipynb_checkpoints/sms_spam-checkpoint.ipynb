```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
```

```python
data =pd.read_csv('SMSSpamCollection.csv',sep = '\t',names = ['label','text'])
```

```python
data
```

```python
# Create binary label for spam/ham
data['b_labels'] = data['label'].map({'spam':1,'ham':0})

# 虽然sklearn可以接受np array以及pandas series，而且接受string作为标签，但是我们自己写模型的时候最好还是将其数值化

Y = data['b_labels'].as_matrix()
```

### 用count做特征

```python
# Create feature from text
Count_Vectorizer = CountVectorizer(decode_error = 'ignore')
X = Count_Vectorizer.fit_transform(data['text'])
```

```python
# 将其向量化之后，是一个及其稀疏的矩阵。行 为 每句话，列 为 每个词， 每个entry里面代表这句话里面，这个词的次数？
X.todense()
```

```python
# train-test-split, 如果用cross validation会得到平均的表现，会比仅仅测试一次要更可靠。
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.33)
```

```python
model = MultinomialNB()
model.fit(Xtrain, Ytrain)
print("train score:", model.score(Xtrain, Ytrain))
print("test score:", model.score(Xtest, Ytest))
```

## 用tf-idf做特征

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(data['text'], Y, test_size=0.33)
tfidf = TfidfVectorizer(decode_error='ignore')
Xtrain = tfidf.fit_transform(Xtrain)
Xtest = tfidf.transform(Xtest)
```

```python
tfidf.idf_
```

```python
model = MultinomialNB()
model.fit(Xtrain, Ytrain)
print("train score:", model.score(Xtrain, Ytrain))
print("test score:", model.score(Xtest, Ytest))
```

### 词云

```python
def visualize(label):
    words = ''
  # 实现对series的遍历
    for msg in data[data['label'] == label]['text']:
        msg = msg.lower()
        words+=msg +' '
    wordcloud = WordCloud(width=600, height=400).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
visualize('spam')
visualize('ham')
```

### 找出分类错误的垃圾邮件

```python
X = tfidf.transform(data['text']) #这个时候tfidf已经被训练集学习了，所以可以用在整个数据集的transform
data['prediction']=model.predict(X)
sneaky_spam = data[(data['prediction'] == 0) & (data['b_labels'] == 1)]['text']
for msg in sneaky_spam:
  print(msg)
```

### 没有误分类的正常邮件

```python
not_actually_spam = data[(data['prediction'] == 1) & (data['b_labels'] == 0)]['text']
for msg in not_actually_spam:
    print(msg)
```
